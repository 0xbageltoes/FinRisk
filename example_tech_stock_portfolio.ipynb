```python
import Portfolio
import datetime as dt
```

```python
endDate = dt.datetime.now()
startDate = endDate - dt.timedelta(days=300)
```

```python
init_cash = 100
my_protfolio = Portfolio.Portfolio(init_cash, startDate, endDate)
```

```python
my_protfolio.add_stock('NVDA', 0.2)
my_protfolio.add_stock('MSFT', 0.2)
my_protfolio.add_stock('AAPL', 0.2)
my_protfolio.add_stock('GOOGL', 0.2)
my_protfolio.add_stock('META', 0.2)
```

```python
my_protfolio.get_portfolio_history()
```

```python
my_protfolio.apply_monte_carlo(100,100)
my_protfolio.get_VaR(alpha=0.05)
my_protfolio.get_conditional_VaR(alpha=0.05)
my_protfolio.plot_monte_carlo()
```
